```python
import numpy as np
import pandas as pd
```

```python
data_test=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv')
data_train=pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv')
```

```python
y_train=data_train['Survived']
```

```python
data_train
```

```python
data_train.drop(labels='Survived',axis=1,inplace=True)
```

```python
data_train.shape
```

```python
train_test=data_train.append(data_test)
```

```python
train_test.isnull().sum()
```

```python
cols_to_drop=["Name","Age","SibSp","Ticket","Cabin","Parch","Embarked"]
train_test.drop(labels=cols_to_drop,axis=1,inplace=True)
```

```python
train_test
```

```python
train_test.isnull().sum()
```

```python
train_test_dummies=pd.get_dummies(train_test,columns=["Sex"])
```

```python
train_test_dummies.shape
```

```python
train_test_dummies.fillna(value=0.0,inplace=True)
```

```python
x_train=train_test_dummies.values[0:891]
x_test=train_test_dummies.values[891:]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train_scale=scaler.fit_transform(x_train)
x_test_scale=scaler.transform(x_test)
```

```python
from sklearn.model_selection import train_test_split
x_train_sub,x_test_sub,y_train_sub,y_test_sub=train_test_split(x_train_scale,y_train,random_state=28)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    ad= AdaBoostClassifier()
    ad.fit(x_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(ad.score(x_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(ad.score(x_test_sub, y_test_sub)))
    print()
```
