```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
# Load the training dataset
train_df = pd.read_csv('train.csv')
```

```python
# Fill missing 'Age' values with the mean age
train_df['Age'].fillna(train_df['Age'].mean(), inplace=True)
```

```python
# Fill missing 'Embarked' values with the mode
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)
```

```python
# Convert 'Sex' to binary (0 for male, 1 for female)
train_df['Sex'] = train_df['Sex'].map({'male': 0, 'female': 1})
```

```python
# Fill missing 'Cabin' values with a placeholder value (e.g., 'Unknown')
train_df['Cabin'].fillna('Unknown', inplace=True)
```

```python
# Fill missing 'Embarked' values with the mode (most frequent value)
mode_embarked = train_df['Embarked'].mode()[0]
train_df['Embarked'].fillna(mode_embarked, inplace=True)
```

```python
# Check for missing values
print(train_df.isnull().sum())
```

```python
# Select features for modeling
features = ['Pclass', 'Sex', 'Age']
```

```python
# Split the data into training and testing sets
X = train_df[features]
y = train_df['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=20)
```

```python
# Build a Naive Bayes model
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
print('Test Data Score',model.score(X_test,y_test))
print('Train data score',model.score(X_train,y_train))
```
